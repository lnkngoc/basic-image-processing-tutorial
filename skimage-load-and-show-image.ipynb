## Edge detection 

```python
from skimage import data, io, filters

image = data.coins()
edges = filters.sobel(image)

io.imshow(edges)
io.show()
```

```python
from skimage import data, io, filters

image = data.coins()
edges = filters.prewitt(image)

io.imshow(edges)
io.show()
```

```python
from skimage import data, io, filters

image = data.coins()
edges = filters.roberts(image)

io.imshow(edges)
io.show()
```

```python
from skimage import data, io, filters

image = data.coins()
edges = filters.laplace(image, 5)

io.imshow(edges)
io.show()
```
