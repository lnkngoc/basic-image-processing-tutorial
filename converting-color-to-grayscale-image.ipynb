# Converting color to grayscale image

useful link: https://www.johndcook.com/blog/2009/08/24/algorithms-convert-color-grayscale/

Three algorithms for converting color to grayscale
- The lightness method
- The average method
- The luminosity method

## The lightness method
The lightness method averages the most prominent and least prominent colors:

- (max(R, G, B) + min(R, G, B))/2

````python
%matplotlib inline 

# %matplotlib inline - activates the inline backend, which renders figures in the notebook as static pngs
````

````python
from skimage import data, io

image = data.coffee()

io.imshow(image)
io.show()
````

````python
print("image shape", image.shape)
print("type of variable", type(image))
print("data type object", image.dtype)
````

````python
print("RGB value ", image[0,0])
print("Maximum RGB value", max(image[0,0]))
print("Minimum RGB value", min(image[0,0]))
````

````python
import numpy as np

gray_image = np.zeros((image.shape[0],image.shape[1]), dtype='uint8')

for row in range(0,(image.shape[0]-1)):
    for col in range(0,(image.shape[1]-1)):
        # (max(R, G, B) + min(R, G, B))/2
        gray_image[row, col] = (max(image[row,col]) + min(image[row,col])) / 2       
````

````python
# using module from scikit-image to show the image
io.imshow(gray_image)
io.show()
````

````python
# using module from matplotlib to show the image
import numpy as np
import matplotlib.pyplot as plt

plt.imshow(gray_image, cmap='gray')
plt.show()
````

````python
print("data type object", gray_image.dtype)
gray_image[0:10,0:10]
````

## The average method
The average method simply averages the values:
- (R + G + B)/3

````python
from skimage import data, io
import numpy as np

image = data.coffee()

print("RGB value", image[0,0])
print("Summation of the RGB value", image[0,0].sum())
````

````python
from skimage import data, io
import numpy as np

image = data.coffee()
gray_image = np.zeros((image.shape[0],image.shape[1]), dtype='uint8')

for row in range(0,(image.shape[0]-1)):
    for col in range(0,(image.shape[1]-1)):
        # (R + G + B)/3
        gray_image[row, col] = image[row,col].sum() / 3

print("data type object", gray_image.dtype)
````

````python
# using module from scikit-image to show the image
io.imshow(gray_image)
io.show()
````

````python
# using module from matplotlib to show the image
import numpy as np
import matplotlib.pyplot as plt

plt.imshow(gray_image, cmap='gray')
plt.show()
````

````python
print("Value of gray image")
print("data type object", gray_image.dtype)
gray_image[0:10,0:10]
````

## The luminosity method
The luminosity method is a more sophisticated version of the average method. It also averages the values, but it forms a weighted average to account for human perception. The formula for luminosity is
- 0.21R + 0.72G + 0.07B

````python
print("Print pixel value", image[0,0])
print("Print Red value", image[0,0][0])
print("Print Green value", image[0,0][1])
print("Print Blue value", image[0,0][2])
````

````python
from skimage import data, io
import numpy as np

image = data.coffee()
gray_image = np.zeros((image.shape[0],image.shape[1]), dtype='uint8')

for row in range(0,(image.shape[0]-1)):
    for col in range(0,(image.shape[1]-1)):
        # 0.21R + 0.72G + 0.07B
        gray_image[row, col] = (0.21*image[row,col][0]) + (0.72*image[row,col][1]) + (0.07*image[row,col][2])

print("data type object", gray_image.dtype)
````

````python
# using module from scikit-image to show the image
io.imshow(gray_image)
io.show()
````

````python
# using module from matplotlib to show the image
import numpy as np
import matplotlib.pyplot as plt

plt.imshow(gray_image, cmap='gray')
plt.show()
````

````python
print("Value of gray image")
print("data type object", gray_image.dtype)
gray_image[0:10,0:10]
````

````python
from skimage import data, io
import numpy as np

image = data.coffee()

# change data type of the gray_image to 'float64'
gray_image = np.zeros((image.shape[0],image.shape[1]), dtype='float64')

for row in range(0,(image.shape[0]-1)):
    for col in range(0,(image.shape[1]-1)):
        # 0.21R + 0.72G + 0.07B
        gray_image[row, col] = (0.21*image[row,col][0]) + (0.72*image[row,col][1]) + (0.07*image[row,col][2])
````

````python
print("Value of gray image")
print("data type object", gray_image.dtype)
gray_image[0:10,0:10]
````

````python
# Normalization the pixel value from 0-255 to 0-1
gray_image = gray_image/255.0

print("data type object", gray_image.dtype)
gray_image[0:10,0:10]
````

## Using scikit-image function to convert color image into gray image

The weights used in this conversion are calibrated for contemporary CRT phosphors:

```Y = 0.2125 R + 0.7154 G + 0.0721``` 

useful link: http://scikit-image.org/docs/dev/api/skimage.color.html#skimage.color.rgb2gray

````python
from skimage import data
from skimage.color import rgb2gray

image = data.coffee()
gray_image = rgb2gray(image)

io.imshow(gray_image)
io.show()
````

````python
print("Value of gray image")
print("data type object", gray_image.dtype)
gray_image[0:10,0:10]
````
